```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as mlp
%matplotlib inline 
```

```python
data = pd.read_csv('C:/Users/User/Desktop/Yukti/Machine Learning/MSIT_ML_CLASS-master/datasets/titanic/train.csv')
```

```python
data.head()
```

```python
data.drop(['Ticket','Cabin'], axis=1, inplace=True)
```

```python
data.info()
```

```python
data.describe()
```

```python
my_filter = (data["Embarked"].isnull())
data[my_filter]
```

```python
data.drop([61,829], axis=0, inplace=True)
```

```python
data.info()
```

```python
my_filt_fare = (data['Fare'] == 0)
data[my_filt_fare]
```

```python
med = data['Fare'].median()
data.Fare.replace(0, med, inplace=True)
data[my_filt_fare]
```

```python
data.loc[data['Parch']+ data['SibSp']==0, 'Family'] = 0
data.loc[data['Parch'] + data['SibSp']!=0, 'Family'] = 1
```

```python
data.head()
```

```python
data.loc[data['Sex'] == 'male', 'Cat_Sex'] = 0
data.loc[data['Sex'] == 'female', 'Cat_Sex'] = 1
```

```python
data.head()
```

```python
my_filt_age = (data["Age"].isnull())
data[my_filt_age]
```

```python
data.drop(['Name','PassengerId'], axis=1, inplace=True)
```

```python
data.head()
```

```python
med_age = data['Age'].median()
print(med_age)
data.Age.replace('NaN', med_age, inplace=True)
data['Age']
```

```python
from sklearn.preprocessing import Imputer 

imp = Imputer(missing_values = 'NaN', strategy = 'median', axis = 0)
imp = imp.fit(data[['Age']])
data['Age'] = imp.transform(data[['Age']]).ravel()
```

```python
data['Age']
```

```python
data1 = pd.get_dummies(data)
```

```python
data1.head()
```

```python
data1.drop(['Cat_Sex', 'Sex_male', 'Embarked_S'], axis=1, inplace=True)
```

```python
data1.head()
```

```python
data1['Cat_Age'] = pd.cut(data1['Age'], bins=[0,18,30,60,80])
```

```python
data1 = pd.get_dummies(data1)
```

```python
data1.head()
```

```python
X = data1.drop(['Survived'], axis=1, inplace=False).values
Y = data1['Survived'].values

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=24)

print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
from sklearn.ensemble import AdaBoostClassifier

clf = AdaBoostClassifier(n_estimators=250, random_state=0)
clf.fit(x_train, y_train)
clf.score(x_test, y_test)
```

```python
clf = AdaBoostClassifier(n_estimators=226, random_state=13)
clf.fit(x_train, y_train)
scr = clf.score(x_test, y_test)
print(scr)
```

```python
from sklearn.ensemble import RandomForestClassifier

#for i in range (200,300):
clf1 = RandomForestClassifier(n_estimators=400, max_depth=1, random_state=13, min_samples_leaf=1)
clf1.fit(x_train, y_train)
scr1 = clf1.score(x_test, y_test)
print(scr)
```

```python
import xgboost as xgb

#for i in range (500,600):
XGB = xgb.XGBClassifier(objective='binary:logistic', n_estimators=521, random_state=13, seed=123)
XGB.fit(x_train, y_train)
scr2 = XGB.score(x_test, y_test)
print(scr2)
```

```python
from sklearn import svm

SVM = svm.SVC(kernel='rbf', random_state=24, gamma=1e2, C=1e9)

SVM.fit(x_train, y_train)
SVM.score(x_test, y_test)
```

```python
from sklearn.linear_model import LogisticRegression

clf3 = LogisticRegression(random_state=13, solver='liblinear', multi_class='auto')
clf3.fit(x_train, y_train)
clf3.score(x_test, y_test)
```
